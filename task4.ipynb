```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('DataSet_task_4.csv')
df.head()
```

```python
X2 = df[['x']]
y2 = df['y']
X2
```

```python
# X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)
```

```python
poly = PolynomialFeatures(degree=10)
# X2_poly_train = poly.fit_transform(X2_train)
# X2_poly_test = poly.fit_transform(X2_test)
X2_poly = poly.fit_transform(X2)
```

```python
model = LinearRegression()
model.fit(X2_poly, y2)
```

```python
y2_pred = model.predict(X2_poly)
y2_pred
```

```python
plt.scatter(X2, y2, color='blue', label='Original Data') 
plt.plot(X2, y2_pred, color='red', label=f'Polynomial Fit (Degree {10})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```

```python
mse_sklearn = mean_squared_error(y2, y2_pred)
print("Sklearn MSE:", mse_sklearn)
```

```python
mse_results = []
degrees = range(1, 21)

for degree in degrees:
    poly = PolynomialFeatures(degree=degree)
    X2_poly = poly.fit_transform(X2)

    model = LinearRegression()
    model.fit(X2_poly, y2)

    y2_pred = model.predict(X2_poly)

    mse = mean_squared_error(y2, y2_pred)
    mse_results.append(mse)
    print(f"Degree {degree}: MSE = {mse}")

# # Plotting the degree vs. MSE
# plt.plot(degrees, mse_results, marker='o')
# plt.xlabel('Degree of Polynomial')
# plt.ylabel('Mean Squared Error (MSE)')
# plt.title('Polynomial Degree vs. MSE')
# plt.grid()
# plt.show()
```
